# Explore Alt-Baltimore GeoJSON Layers

```
import geopandas as gpd
import matplotlib.pyplot as plt

# --- Load GeoJSON layers ---
points = gpd.read_file("../data/points/alt_baltimore_points.geojson")
lines = gpd.read_file("../data/lines/alt_baltimore_lines.geojson")
polygons = gpd.read_file("../data/polygons/alt_baltimore_polygons.geojson")

# --- Display basic info ---
print("Points:", len(points))
print("Lines:", len(lines))
print("Polygons:", len(polygons))

display(points.head())
display(lines.head())
display(polygons.head())

# --- Plot each layer individually ---
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

polygons.plot(ax=axs[0], color='lightblue', edgecolor='black')
axs[0].set_title("Polygons")

lines.plot(ax=axs[1], color='red')
axs[1].set_title("Lines")

points.plot(ax=axs[2], color='green', markersize=20)
axs[2].set_title("Points")

plt.tight_layout()
plt.show()

# --- Combined overlay map ---
fig, ax = plt.subplots(figsize=(10, 10))
polygons.plot(ax=ax, color='lightblue', edgecolor='black')
lines.plot(ax=ax, color='red')
points.plot(ax=ax, color='green', markersize=25)

ax.set_title("Alt-Baltimore â€” Combined Layers")
plt.show()
```
